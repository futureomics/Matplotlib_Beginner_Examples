### Basic Line Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Line Plot')
plt.show()
```

### Multiple Lines

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 3, 4, 5, 6]
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = x + 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('Multiple Lines')
plt.show()
```

### Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y, color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

### Bar Plot

```python
import matplotlib.pyplot as plt
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 2, 5, 4]
plt.bar(x, y, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

### Horizontal Bar Plot

```python
import matplotlib.pyplot as plt
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 2, 5, 4]
plt.barh(x, y, color='purple')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
sizes = [25, 35, 15, 25]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.xlabel('Data sets')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

### Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.violinplot(data=data)
plt.xlabel('Data sets')
plt.ylabel('Values')
plt.title('Violin Plot')
plt.show()
```

### Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = np.random.rand(10, 10)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

### Subplots

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-')
plt.title('Line Plot')
plt.subplot(1, 2, 2)
plt.bar(x, y, color='blue')
plt.title('Bar Plot')
plt.show()
```

### Filled Area Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [0, 2, 4, 6, 8]
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.plot(x, y1, color='blue', alpha=0.6, linewidth=2)
plt.plot(x, y2, color='red', alpha=0.6, linewidth=2)
plt.title('Filled Area Plot')
plt.show()
```

### 3D Line Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 20, 100)
x = np.sin(t)
y = np.cos(t)
z = t
ax.plot(x, y, z, label='3D Line')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

### 3D Scatter Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)
ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

### Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r
plt.polar(theta, r)
plt.title('Polar Plot')
plt.show()
```

### Logarithmic Scale

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.1, 10, 100)
y = np.exp(x)
plt.plot(x, y)
plt.yscale('log')
plt.xlabel('x')
plt.ylabel('log scale')
plt.title('Logarithmic Scale Plot')
plt.show()
```

### Annotations

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.annotate('Peak', xy=(5, 25), xytext=(3, 20),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Plot with Annotations')
plt.show()
```

### Adding Grid

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.grid(True)
plt.title('Plot with Grid')
plt.show()
```

### Custom Ticks

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([0, 5, 10, 15, 20, 25])
plt.title('Plot with Custom Ticks')
plt.show()
```

### Shaded Error Bars

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
error = 0.2
plt.plot(x, y, label='sin(x)')
plt.fill_between(x, y - error, y + error, alpha=0.2, color='orange')
plt.title('Shaded Error Bars')
plt.show()
```
